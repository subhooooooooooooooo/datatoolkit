1.What is NumPy, and why is it widely used in Python?
A: NumPy is a library for numerical computing in Python. It provides support for multi-dimensional arrays and high-level mathematical functions, making it efficient for data manipulation and scientific computation.

2.How does broadcasting work in NumPy?
A: Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes by automatically expanding the smaller array’s dimensions to match the larger one.

3 What is a Pandas DataFrame
A Pandas DataFrame is a two-dimensional, tabular data structure in Python provided by the pandas library. It is one of the most widely used tools for data analysis and manipulation.

4, Explain the use of the groupby() method in Pandas
The .groupby() method in Pandas is used to group data based on the values in one or more columns, and then perform operations

5 Why is Seaborn preferred for statistical visualizations
Seaborn is preferred for statistical visualizations in Python because it builds on Matplotlib and adds a higher-level, easier-to-use API that’s specifically designed for statistical plotting.

6 What are the differences between NumPy arrays and Python lists?
 NumPy arrays are faster, support vectorized operations, use less memory, and require homogeneous data types, whereas Python lists can hold mixed types and are less efficient for numerical tasks.

7 What is a heatmap, and when should it be used)
A heatmap is a graphical representation of data where individual values contained in a matrix are represented using color gradients

8 What does the term “vectorized operation” mean in NumPy?
 It refers to applying operations directly on entire arrays without explicit loops, resulting in faster and more readable code.

Matplotlib and Plotly are both powerful Python libraries for data visualization, but they differ significantly in their design goals, interactivity, and ease of use

11 What is the role of Seaborn’s pairplot() function)
 The pairplot() function creates pairwise scatter plots and histograms for numerical variables, helping to quickly visualize relationships between multiple features.

12 What is the purpose of the describe() function in Pandas)
The describe() function provides summary statistics for numerical columns, including count, mean, standard deviation, min, max, and quartiles.

```python
# 4  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap)
# Step 1: Import libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load dataset
df = sns.load_dataset("titanic")

# Step 3: Select numeric columns only (correlation works with numerical data)
numeric_df = df.select_dtypes(include=["float64", "int64"])

# Step 4: Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Step 5: Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix Heatmap - Titanic Dataset")
plt.show()
```
